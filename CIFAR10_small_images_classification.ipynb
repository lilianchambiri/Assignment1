```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
print('x_test:', x_test.shape)
print('y_test', y_test.shape)
```

```python
x_train_scaled = x_train/255
x_test_scaled = x_test/255
```

```python
print('x_train_scaled:', x_train_scaled.shape)
print('y_train:', y_train.shape)
print('x_test_scaled:', x_test_scaled.shape)
print('y_test', y_test.shape)
```

```python
# flattening one image
x_train_scaled[0].reshape((28*28,)).shape
```

```python
x_train_flatten = x_train_scaled.reshape((60000,28*28))
x_test_flatten = x_test_scaled.reshape((10000,28*28))
```

```python
print('x_train_flatten:', x_train_flatten.shape)
print('x_test_flatten:', x_test_flatten.shape)
```

```python
# defining the model
model = keras.Sequential([
    keras.layers.Dense(units=3026, activation='relu', kernel_regularizer=keras.regularizers.L2(), input_shape=(784,)),
    keras.layers.Dense(units=1023, activation='relu'),
    keras.layers.Dense(units=512, activation='relu'),
    keras.layers.Dense(units=256, activation='relu'),
    keras.layers.Dense(units=10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    x_train_flatten,
    y_train,
    epochs=20,
    validation_data=(x_test_flatten,y_test))
```

```python
model.save('current_model.h5')
```

```python
# How to upload a model
model_upload = keras.models.load_model('/content/current_model.h5')
```

```python
model_upload.summary()
```

```python
model_upload.evaluate(x_train_flatten, y_train)
```

```python
model_upload.evaluate(x_test_flatten, y_test)
```

```python
y_train_pred = model_upload.predict(x_train_flatten)
y_test_pred = model_upload.predict(x_test_flatten)
```

```python
y_train_pred_final = [np.argmax(i) for i in y_train_pred]
y_test_pred_final = [np.argmax(i) for i in y_test_pred]
```
