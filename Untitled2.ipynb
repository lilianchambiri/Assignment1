```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("heart.csv")
data.head()
```

```python
data.shape
```

```python
data['HeartDisease'].value_counts()
```

```python
# Using map to convert categorical values to numerial values
data['Sex'] = data['Sex'].map({'M': 1, 'F':0})
data.head()
```

```python
data['ExerciseAngina'] = data['ExerciseAngina'].apply(lambda x: 1 if x=='Y' else 0)
```

```python
data.head()
```

```python
data.RestingECG.value_counts()
```

```python
data[['Normal', 'ST']] = pd.get_dummies(data['RestingECG'], drop_first=True)
data.head()
```

```python
data.drop("RestingECG", axis=1, inplace=True)
```

```python
data.head(3)
```

```python
data.ST_Slope.value_counts()
```

```python
data[['Flat', 'Up']] = pd.get_dummies(data['ST_Slope'], drop_first=True)
data.drop('ST_Slope', axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
data['Age'].min()
```

```python
data['Age'].max()
```

```python
def scale(x):
  return (x - 28)/(77 - 28)
```

```python
data['Age'].apply(scale)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
scale.fit(data[['Age', 'RestingBP', 'Cholesterol', 'MaxHR']])
```

```python
data[['Age', 'RestingBP', 'Cholesterol', 'MaxHR']] = scale.transform(data[['Age', 'RestingBP', 'Cholesterol', 'MaxHR']])
```

```python
# alternatively we use fit_tansform() to perform both fit() and transform()
# scale.fit_transform()
```

```python
data.head(2)
```

```python
data.shape
```

```python
#Dividing the data into test and train data
from sklearn.model_selection import train_test_split
```

```python
x = data.drop("HeartDisease", axis=1)
y = data["HeartDisease"]
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.25,random_state=101, stratify=y)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Building the model
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(units=1,activation="sigmoid",input_shape=(12,))
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
model.fit(x=x_train,y=y_train, epochs=50,validation_data=(x_test,y_test))
```

```python
model.fit(x=x_train,y=y_train, epochs=80,validation_data=(x_test,y_test),initial_epoch=50)
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
```

```python
print(y_pred_train.shape)
print(y_train.shape)
print(y_pred_test.shape)
print(y_test.shape)
```

```python
y_pred_train=y_pred_train.reshape(688,)
y_pred_test=y_pred_test.reshape(230,)
```

```python
print(y_pred_train.shape)
print(y_train.shape)
print(y_pred_test.shape)
print(y_test.shape)
```

```python
print("Report for training data: ")
print(classification_report(y_train, y_pred_train_final))
print("Report for testing data: ")
print(classification_report(y_test, y_pred_test_final))
```
